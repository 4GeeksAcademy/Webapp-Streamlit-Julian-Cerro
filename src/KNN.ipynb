# K-Nearest Neighbour

## Step 1: Loading the dataset

```python
import pandas as pd

movies = pd.read_csv("../data/raw/tmdb_5000_movies.csv")
movies.head()
```

```python
credits = pd.read_csv("../data/raw/tmdb_5000_credits.csv")
credits.head()
```

## Step 2: Creation of the database

```python
import sqlite3

conn = sqlite3.connect("../data/movies_database.db")

movies.to_sql("movies_table", conn, if_exists = "replace", index = False)
credits.to_sql("credits_table", conn, if_exists = "replace", index = False)
```

```python
#Merging tables for creating a new DataFrame
query = """
    SELECT credits_table.movie_id, movies_table.title, movies_table.overview, 
           movies_table.genres, movies_table.keywords, 
           credits_table.cast, credits_table.crew
    FROM movies_table
    INNER JOIN credits_table
    ON movies_table.title = credits_table.title;
"""
total_data = pd.read_sql_query(query, conn)

conn.close()

total_data.head()
```

## Step 3: Transform the data

```python
import json
import pandas as pd

#Parse JSON safely
def parse_json_column(column_data):
    try:
        return json.loads(column_data.strip())
    except (AttributeError, json.JSONDecodeError):
        return None
```

```python
#Extract "name" fields from JSON data
def extract_names(json_data):
    if json_data:
        return [item['name'] for item in json_data if 'name' in item]
    return []
```

```python
#Extract the first 3 cast members
def extract_top_3_cast(json_data):
    if json_data:
        return [item['name'] for item in json_data[:3] if 'name' in item]
    return []
```

```python

#director's name from "crew"
def extract_director(json_data):
    if json_data:
        for crew_member in json_data:
            if crew_member.get('job') == 'Director':
                return crew_member['name']
    return None
```

```python
def remove_spaces_from_list(text_list):
    try:
        #Iterate and replace spaces
        return [word.replace(" ", "") for word in text_list]
    except (TypeError, AttributeError):
        #Else, .replace replaces directly
        try:
            return text_list.replace(" ", "")
        except AttributeError:
            #If otherwise its not a string, return original content
            return text_list
```

```python
total_data["overview"] = total_data["overview"].apply(lambda x: [str(x)] if pd.notna(x) else [])

#Parse JSON and extract relevant fields
total_data["genres"] = total_data["genres"].apply(lambda x: extract_names(parse_json_column(x)))
total_data["keywords"] = total_data["keywords"].apply(lambda x: extract_names(parse_json_column(x)))
total_data["cast"] = total_data["cast"].apply(lambda x: extract_top_3_cast(parse_json_column(x)))
total_data["crew"] = total_data["crew"].apply(lambda x: extract_director(parse_json_column(x)))
```

```python
#Remove spaces
total_data["genres"] = total_data["genres"].apply(remove_spaces_from_list)
total_data["keywords"] = total_data["keywords"].apply(remove_spaces_from_list)
total_data["cast"] = total_data["cast"].apply(remove_spaces_from_list)
total_data["crew"] = total_data["crew"].apply(remove_spaces_from_list)
```

```python
#New tags Column
total_data["tags"] = (
    total_data["overview"] +
    total_data["genres"] +
    total_data["keywords"] +
    total_data["cast"] +
    total_data["crew"].apply(lambda x: [x] if x else [])
)
```

```python
#Convert elements to strings
total_data["tags"] = total_data["tags"].apply(lambda x: [str(item) for item in x if item is not None])

#Join all elements in the tags column into a single string
total_data["tags"] = total_data["tags"].apply(lambda x: " ".join(x))

total_data.drop(columns=["genres", "keywords", "cast", "crew", "overview"], inplace=True)

print(total_data.iloc[0].tags)
```

```python
total_data.to_csv("../data/processed/clean_data.csv", index = False)

conn = sqlite3.connect("../data/movies_database.db")

movies.to_sql("clean_movies_data", conn, if_exists = "replace", index = False)
```

## K-NN Model

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors

print(total_data.head()) 
```

```python
#Convert 'tags' into TF-IDF vectors
vectorizer = TfidfVectorizer()  
tfidf_matrix = vectorizer.fit_transform(total_data['tags'])  # Transform tags into TF-IDF matrix
```

```python
#Initialize and Train the KNN Model
knn = NearestNeighbors(metric='cosine', algorithm='brute') 
knn.fit(tfidf_matrix)  
```

```python
def recommend(movie_title, num_recommendations=5):
    try:
        movie_index = total_data[total_data['title'] == movie_title].index[0]
        
       #Indexes
        _, indices = knn.kneighbors(tfidf_matrix[movie_index], n_neighbors=num_recommendations + 1)
        
        print(f"Movies similar to '{movie_title}':")
        for i in indices[0][1:]: #skip first index
            print(total_data.iloc[i]['title'])
    except IndexError:
        print(f"Movie '{movie_title}' not found in the dataset!")

recommend("Spectre")
```
